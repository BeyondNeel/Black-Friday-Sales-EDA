```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
#merge both datasets
df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# df['Gender'] = pd.get_dummies(df['Gender'],drop_first=1)
# Hndling categorical feature Gender

df['Gender'] = df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
# Handling categorical feature Age
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
 # Fixing the feature city_category
 df_city = pd.get_dummies(df['City_Category'],drop_first=True)
 df_city.head()
```

```python
df = pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Missing values
df.isnull().sum()
```

```python
#replace missing values
df.Age.unique()
```

```python
df.Product_Category_2.unique()
```

```python
df.Product_Category_3.unique()
```

```python
#replace the missing values with mode
df.Age = df.Age.fillna(df.Age.mode()[0])
```

```python
df.Age.isnull().sum()
```

```python
df.Product_Category_2 = df.Product_Category_2.fillna(df.Product_Category_2.mode()[0])
```

```python
df.Product_Category_2.isnull().sum()
```

```python
df.Product_Category_3 = df.Product_Category_3.fillna(df.Product_Category_3.mode()[0])
```

```python
df.Product_Category_3.isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.extract('(\d+)').astype(float)
```

```python
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```
